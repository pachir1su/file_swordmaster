```python
import os
import glob
from pypdf import PdfReader, PdfWriter

print("м—…лЎңл“ңлҗң PDF нҢҢмқјмқ„ м°ҫлҠ” мӨ‘...")
pdf_files = glob.glob("/content/*.pdf")

if not pdf_files :
    print("PDF нҢҢмқјмқ„ м°ҫмқ„ мҲҳ м—ҶмҠөлӢҲлӢӨ. нҢҢмқј м—…лЎңл“ңк°Җ мҷ„лЈҢлҗҳм—ҲлҠ”м§Җ нҷ•мқён•ң нӣ„ лӢӨмӢң мӢӨн–үн•ҙ мЈјм„ёмҡ”.")
else :
    target_pdf = pdf_files[0]
    file_name, ext = os.path.splitext(os.path.basename(target_pdf))
    print(f"мһҗлҸҷ мқёмӢқлҗң нҢҢмқј: '{file_name}{ext}'\n")

    try:
        reader = PdfReader(target_pdf)
        total_pages = len(reader.pages)
        print(f"н•ҙлӢ№ PDFмқҳ мҙқ нҺҳмқҙм§Җ мҲҳ: {total_pages}нҺҳмқҙм§Җ")
        print("-" * 40)

        while True :
            try:
                start_input = input(f"м¶”м¶ңмқ„ мӢңмһ‘н•  нҺҳмқҙм§Җ лІҲнҳёлҘј мһ…л Ҙн•ҳм„ёмҡ” (1 ~ {total_pages}) : ")
                end_input = input(f"м¶”м¶ңмқ„ лҒқлӮј нҺҳмқҙм§Җ лІҲнҳёлҘј мһ…л Ҙн•ҳм„ёмҡ” ({start_input} ~ {total_pages}) : ")

                start_page = int(start_input)
                end_page = int(end_input)

                if 1 <= start_page <= end_page <= total_pages :
                    break
                else :
                    print("нҺҳмқҙм§Җ лІ”мң„к°Җ мһҳлӘ»лҗҳм—ҲмҠөлӢҲлӢӨ. лӢӨмӢң мһ…л Ҙн•ҙ мЈјм„ёмҡ”.\n")
            except ValueError :
                print("мҲ«мһҗл§Ң мһ…л Ҙн•ҙ мЈјм„ёмҡ”.\n")

        print("-" * 40)
        print(f"{start_page}нҺҳмқҙм§Җл¶Җн„° {end_page}нҺҳмқҙм§Җк№Ңм§Җ м¶”м¶ңмқ„ м§„н–үн•©лӢҲлӢӨ.")

        writer = PdfWriter()
        for i in range(start_page - 1, end_page) :
            writer.add_page(reader.pages[i])

        output_name = f"{file_name}_{start_page}to{end_page}{ext}"
        output_path = f"/content/{output_name}"

        with open(output_path, "wb") as output_file :
            writer.write(output_file)

        print(f"мһ‘м—… мҷ„лЈҢ. '{output_name}' нҢҢмқјмқҙ м ҖмһҘлҗҳм—ҲмҠөлӢҲлӢӨ.")
        print("мўҢмёЎ нҸҙлҚ” нғӯмқ„ мғҲлЎңкі м№Ён•ҳм—¬ кІ°кіјл¬јмқ„ нҷ•мқён•ҳмӢӯмӢңмҳӨ.")

    except Exception as e :
        print(f"мһ‘м—… мӨ‘ мҳӨлҘҳк°Җ л°ңмғқн–ҲмҠөлӢҲлӢӨ: {e}")
```
