```python
import os
import glob
from pypdf import PdfReader, PdfWriter

print("업로드된 PDF 파일을 찾는 중...")
pdf_files = glob.glob("/content/*.pdf")

if not pdf_files :
    print("PDF 파일을 찾을 수 없습니다. 파일 업로드가 완료되었는지 확인한 후 다시 실행해 주세요.")
else :
    target_pdf = pdf_files[0]
    file_name, ext = os.path.splitext(os.path.basename(target_pdf))
    print(f"자동 인식된 파일: '{file_name}{ext}'\n")

    try:
        reader = PdfReader(target_pdf)
        total_pages = len(reader.pages)
        print(f"해당 PDF의 총 페이지 수: {total_pages}페이지")
        print("-" * 40)

        while True :
            try:
                start_input = input(f"추출을 시작할 페이지 번호를 입력하세요 (1 ~ {total_pages}) : ")
                end_input = input(f"추출을 끝낼 페이지 번호를 입력하세요 ({start_input} ~ {total_pages}) : ")

                start_page = int(start_input)
                end_page = int(end_input)

                if 1 <= start_page <= end_page <= total_pages :
                    break
                else :
                    print("페이지 범위가 잘못되었습니다. 다시 입력해 주세요.\n")
            except ValueError :
                print("숫자만 입력해 주세요.\n")

        print("-" * 40)
        print(f"{start_page}페이지부터 {end_page}페이지까지 추출을 진행합니다.")

        writer = PdfWriter()
        for i in range(start_page - 1, end_page) :
            writer.add_page(reader.pages[i])

        output_name = f"{file_name}_{start_page}to{end_page}{ext}"
        output_path = f"/content/{output_name}"

        with open(output_path, "wb") as output_file :
            writer.write(output_file)

        print(f"작업 완료. '{output_name}' 파일이 저장되었습니다.")
        print("좌측 폴더 탭을 새로고침하여 결과물을 확인하십시오.")

    except Exception as e :
        print(f"작업 중 오류가 발생했습니다: {e}")
```
