```python
import os
import glob
from pydub import AudioSegment

# 1. 원하는 분할 개수(n)만 설정하세요.
n_splits = 3

print("업로드된 오디오 파일을 찾는 중...")

# /content 폴더 안의 m4a 또는 mp3 파일을 자동으로 찾습니다.
target_files = glob.glob("/content/*.m4a") + glob.glob("/content/*.mp3")

if not target_files:
    print("파일을 찾을 수 없습니다. 왼쪽 창에서 파일 업로드가 '완전히' 끝났는지 확인한 후 다시 실행해 주세요!")
else:
    # 가장 먼저 찾은 파일을 자동으로 선택
    file_path = target_files[0]
    file_name, ext = os.path.splitext(os.path.basename(file_path))

    print(f"자동 인식된 파일: '{file_name}{ext}'")
    print("불러오는 중입니다. 잠시만 기다려주세요...")

    # 2. 오디오 파일 불러오기 및 길이 확인
    audio = AudioSegment.from_file(file_path)
    total_length_ms = len(audio)

    print(f"총 길이는 약 {total_length_ms // 1000}초입니다. {n_splits}등분 분할을 시작합니다.")

    # 3. 파일 N등분 쪼개기
    chunk_length_ms = total_length_ms // n_splits
    chunks = []

    for i in range(n_splits):
        start_time = i * chunk_length_ms
        end_time = (i + 1) * chunk_length_ms if i < n_splits - 1 else total_length_ms
        chunks.append(audio[start_time:end_time])

    # 4. 쪼갠 파일들을 저장
    for i, chunk in enumerate(chunks):
        output_name = f"{file_name}_part{i+1}{ext}"
        export_format = "ipod" if ext.lower() == ".m4a" else ext.replace('.', '')

        # 파일 저장
        chunk.export(f"/content/{output_name}", format=export_format)
        print(f"분할 저장 완료 : {output_name}")

    print("작업이 끝났습니다! 왼쪽 폴더 탭에서 새로고침을 눌러 확인하세요.")
```
